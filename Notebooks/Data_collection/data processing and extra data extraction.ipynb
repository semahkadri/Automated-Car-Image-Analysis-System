```python
df = pd.read_csv('cars_brands_description.csv')
```

```python
df["car type"] = df["url"].str.split('/').str[3]
df['car name'] = df['url'].str.split('/').str[3] + ' ' + df['url'].str.split('/').str[4]
df['year'] = df['url'].str.split('/').str[5]
```

```python
#remove the first 2 columns
df = df.drop(df.columns[[0, 1]], axis=1)
```

```python
#reorder the columns
df = df[['car type', 'car name', 'year', 'url', 'Pros', 'Cons','Full Description']]
```

```python
#count each unique value in year column
df['year'].value_counts()
```

```python
#count nan values
df['year'].isnull().sum()
```

```python
# if value in year is '' make it NaN
df['year'] = df['year'].replace('', np.nan)
```

```python
df['year1']=df["Full Description"].str.extract("(\d{4})")
df['year'] = df['year'].fillna(df['year1'])
df=df.drop(['year1'],axis=1)
```

```python
s = HTMLSession()
def get_website(url):
    response = s.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    return soup

def get_rating(soup):
    rating = soup.find('span', class_='average-user-rating')
    if rating is None:
        return None
    return rating.text

def get_princing(soup):
    princigDiv = soup.find('div', class_='intro-trims-content')
    if princigDiv is None:
        return None, None
    pricing = princigDiv.find('td', class_='text-right align-top')
    if pricing is None:
        return None, None
    li = pricing.text.split('-')
    if len(li) == 1:
        return li[0], li[0]
    return li[0], li[1]

def get_car_features(soup):
    features = soup.find('div', class_='features-lists')
    if features is None:
        return None
    return features.text

def get_car_specs(soup):
    specs = soup.find('div', class_='features-specs')
    #print("specs", specs)
    if specs is None:
        return None, None, None, None
    dl = specs.find_all('dl')
    #print("dl", dl)
    if dl is None:
        return None, None, None, None
    # iterate through each dl and check df text if it exists extract all dd text from dl
    base_msrp = ''
    engine = ''
    Drivetrain = ''
    Dimenstions = ''
    for ele in dl:
        if ele.dt.text == 'Base MSRP' or ele.dt.text == 'Total MSRP':
            # get list of dd elements from ele
            base_msrpList = ele.find_all('dd')
            #ieterate through each dd element and get text
            for ele1 in base_msrpList:
                base_msrp += ele1.text + ' \n '
        if ele.dt.text == 'Engine':
            engineList = ele.find_all('dd')
            for ele1 in engineList:
                engine += ele1.text + ' \n '
        if ele.dt.text == 'Drivetrain':
            DrivetrainList = ele.find_all('dd')
            for ele1 in DrivetrainList:
                Drivetrain += ele1.text + ' \n '
        if ele.dt.text == 'Dimensions':
            DimenstionsList = ele.find_all('dd')
            for ele1 in DimenstionsList:
                Dimenstions += ele1.text + ' \n '

    return base_msrp, engine, Drivetrain, Dimenstions

```

```python
#iterate over url column
from tqdm import tqdm
for index, row in tqdm(df.iterrows()):
    url = row['url']
    #sleep(random.uniform(0, 1))
    soup = get_website(url)
    rating = get_rating(soup)
    priceMin, priceMax = get_princing(soup)
    base_msrp, engine, Drivetrain, Dimensions = get_car_specs(soup)
    #print((base_msrp, engine, Drivetrain, Dimensions))
    df.at[index, 'rating'] = rating
    df.at[index, 'current price Min'] = priceMin
    df.at[index, 'current price Max'] = priceMax
    df.at[index, 'base msrp'] = base_msrp
    df.at[index, 'engine'] = engine
    df.at[index, 'drivetrain'] = Drivetrain
    df.at[index, 'dimensions'] = Dimensions
```

```python
#save to csv
df.to_csv('carsSemiProcessed.csv', index=False)
```
